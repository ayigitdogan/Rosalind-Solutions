```python
# Importing the "json" library to read codon table

import json
```

```python
# Function definition

def rna_to_protein(rna_string):

    with open('Data/rna_codon_table.txt') as f:
        rna_codon_table = json.loads(f.read())

    protein = ""

    for i in range(0, len(rna_string) - 2, 3):

        codon = rna_string[i:i+3]
        amino_acid = rna_codon_table[codon]

        # Stopping condition
        if amino_acid == "Stop":
            break

        protein += rna_codon_table[codon]

    return(protein)
```

```python
with open('Data/rna_codon_table.txt') as f:
        rna_codon_table = json.loads(f.read())
```

```python
# Reading the input file

with open('Data/008_rosalind_prot.txt') as data:
    rna_string = data.read().splitlines()[0]
```

```python
# Applying the function

rna_to_protein(rna_string)
```
