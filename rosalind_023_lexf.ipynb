# Enumerating k-mers Lexicographically

```python
# Function definition

def enum_lex(alphabet, k):

    n = len(alphabet)
    size = n**k
    sets = []

    for col in reversed(range(0, k)):

        set = sorted(alphabet*(n**col))*int(size/((n**col)*n))
        sets.append(set)

    sets = list(zip(*sets))
    sets = ["".join(row) for row in sets]

    return(sets)
```

```python
# Reading the input data

with open('Data/023_rosalind_lexf.txt') as d:
    data = d.read().splitlines()
    alphabet = list(data[0].replace(" ", ""))
    k = int(data[1])
```

```python
# Applying the function

results = enum_lex(alphabet, k)
for result in results:
    print(result)
```
